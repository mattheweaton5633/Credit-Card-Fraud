Prompt: Now create another countplot, make x = "used_pin_number", hue is "fraud" and palett is green and red. Title it 'Transactions with PIN vs Fraudulent Transactions'. Make the x axis label "Used PIN" and y label "Count". For the legend, the title is "Fraud", and the labels are "Non-Fraudulent" and "Fraudulent".

```python
# Create a count plot comparing PIN usage and fraud
sns.countplot(
    x="used_pin_number",
    hue="fraud",
    data=transaction_data_cleaned,
    palette={0: "green", 1: "red"}
)

# Add title and axis labels
plt.title("Transactions with PIN vs Fraudulent Transactions")
plt.xlabel("Used PIN")
plt.ylabel("Count")

# Customize legend
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Show the plot
plt.show()
```

Prompt: First, group 'fraud' by 'used_pin_number', count the values, unstack, and multiply by 100. Call this 'df_pin_fraud'. Then create a bar plot like before titled "Percentage of Fraudulent Transactions with PIN Usage".

```python

# 1. Group 'fraud' by 'used_pin_number', count values, unstack, and convert to percentages
df_pin_fraud = (
    transaction_data_cleaned.groupby('used_pin_number')['fraud']
    .value_counts(normalize=True)
    .unstack(fill_value=0) * 100
)

# 2. Create stacked bar plot
df_pin_fraud.plot(
    kind='bar',
    stacked=True,
    color={0: 'green', 1: 'red'},
    figsize=(7,5)
)

# 3. Add title, labels, and legend
plt.title("Percentage of Fraudulent Transactions with PIN Usage")
plt.xlabel("Used_PIN")
plt.ylabel("Percentage (%)")
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# 4. Show the plot
plt.show()

```

Prompt: Now create a countplot with 'used_chip' on the x axis and the hue is 'fraud' like earlier and title it "Transactions with Chip vs Fraudulent Transactions". Then create another percentage plot like earlier titled Percentage of Fraudulent Transactions with Chip Usage" with 'used_chip'.

Prompt: Now create a scatterplot titled "Fraudulent vs Non-Fraudulent Transactions: Distance from Last Transaction vs Ratio to Median Purchase Price" with "distance_from_last_transaction" on the x and "ratio_to_median_purchase_price" on the y.
